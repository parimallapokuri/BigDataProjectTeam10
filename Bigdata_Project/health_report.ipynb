```python
health = spark.read.format("csv").option("header", "true").load("health.csv")
health.createOrReplaceTempView("health")
health.show()
```

### Goal-1: 
Percentage of men married before 21 years in Rural and Urban areas with respect to state 

```python
# Maximum Percentage of men married before 21 years in Rural accroding to state
spark.sql('select State_Name,max(EE_Currently_Married_Men_Aged_25_29_Years_Married_Before_Legal_Age_21_Years_Rural) as Married_man_before21_rural from health group by State_Name').show()
```

```python
# Min Percentage of men married before 21 years in Urban area accroding to state
spark.sql('select State_Name,min(EE_Currently_Married_Men_Aged_25_29_Years_Married_Before_Legal_Age_21_Years_Urban) as Married_man_before21_urban from health  group by State_Name').show()
```

### Goal-2
The total average population for each state wise

```python

spark.sql('select State_Name,avg(AA_Population_Total) as Average_Population from health  group by State_Name order by 1').show()
```

### Goal3
District-wise number of children in rural and urban areas (Ages 12 to 23)

```python
spark.sql('select State_Name,State_District_Name, SUM(AA_Children_12_23_Months_Rural) as Rural_Children_Population from health group by State_Name,State_District_Name order by  Rural_Children_Population desc').show()
```

```python
spark.sql('select State_Name,State_District_Name, SUM(AA_Children_12_23_Months_Urban) as Urban_Children_Population from health group by State_Name,State_District_Name order by Urban_Children_Population').show()
```

### Goal-4
Total number of children currently attending school between 6 to 17 years

```python
# FF_Children_Currently_Attending_School_Age_6_17_Years_Person_Total
# AA_Children_12_23_Months_Total

spark.sql('select State_District_Name,AA_Children_12_23_Months_Total as Total_Child_Population,round(((FF_Children_Currently_Attending_School_Age_6_17_Years_Person_Total*AA_Children_12_23_Months_Total)/100)) as Total_Children_Attending_School,(AA_Children_12_23_Months_Total-round(((FF_Children_Currently_Attending_School_Age_6_17_Years_Person_Total*AA_Children_12_23_Months_Total)/100))) as Total_Child_not_attending from health order by 1').show()
```

### Goal-5
Average Abortion month by women in rural and urban areas

```python
#Rural Areas

spark.sql('select State_Name,round(avg(MM_Average_Month_Of_Pregnancy_At_The_Time_Of_Abortion_Rural),1) as avg_abortion_month_rural from health  group by State_Name order by 1').show()
```

```python
#Urban Areas

spark.sql('select State_Name,round(avg(MM_Average_Month_Of_Pregnancy_At_The_Time_Of_Abortion_Urban),1) as avg_abortion_month_urban from health  group by State_Name order by 1').show()
```

### Goal-6
Percentage of women who are aware of signs of pneumonia

```python
# XX_Women_Who_Are_Aware_Of_Danger_Signs_Of_Ari_Pneumonia_Total
spark.sql('select State_Name,round(avg(XX_Women_Who_Are_Aware_Of_Danger_Signs_Of_Ari_Pneumonia_Total),2) as avg_percentage_women_aware_about_pneumonia from health  group by State_Name order by 1').show()
```

### Goal-7
Which state has the maximum and minimum death rates

```python
spark.sql('select State_Name,max(YY_Crude_Death_Rate_Cdr_Total_Person) as Crude_Death_Rate from health group by State_Name').show()
```

### Goal-8
Percentage of women married before 18 years in Rural and Urban areas

```python
spark.sql('select State_Name,round(avg(EE_Marriages_Among_Females_Below_Legal_Age_18_Years_Total),2) as women_married_b4_18 from health group by State_Name order by 2').show()
```

### Goal-9
Which state has max and min birth rates

```python
# spark.sql('select round(max(LL_Crude_Birth_Rate_Cbr_Total),2) as Max_birth_rate from health order by 1').show()

# SELECT State_Name, LL_Crude_Birth_Rate_Cbr_Total
# FROM health
# WHERE (LL_Crude_Birth_Rate_Cbr_Total) IN (
#     SELECT MAX(LL_Crude_Birth_Rate_Cbr_Total)
#     FROM health
#     GROUP BY State_Name
# )
spark.sql('SELECT State_Name, LL_Crude_Birth_Rate_Cbr_Total FROM health WHERE (State_Name,LL_Crude_Birth_Rate_Cbr_Total) IN (SELECT State_Name, MAX(LL_Crude_Birth_Rate_Cbr_Total) FROM health GROUP BY State_Name)').show()
```

### Goal-10
Usage of copper in both rural and urban areas

```python
# NN_Current_Usage_Copper_T_Iud_Rural
# NN_Current_Usage_Copper_T_Iud_Urban

spark.sql('select State_Name,round(avg(NN_Current_Usage_Copper_T_Iud_Rural),2) as copper_usage_rural from health group by State_Name order by 2').show()
```
