```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('health.csv')
df
```

```python
print("checking for null values")
print(df.isnull().sum())
```

```python
# count null values in each column
null_cols = df.isnull().sum()

# drop null columns
df = df.drop(columns=null_cols[null_cols > 0].index)

print(df)
```

```python
# count null values in each column
null_cols = df.isnull().sum()

# display null columns and their count
null_cols = null_cols[null_cols > 0] # select columns with null values
print("Null values in columns:")
print(null_cols)

print("checking for null values")
print(df.isnull().sum())
```

```python
#creating train and test sets
from sklearn.model_selection import train_test_split
train_set, test_set = train_test_split(df, test_size=0.25, random_state=531)
print(len(train_set), len(test_set))
print(train_set.head())
print(test_set.head())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC

import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeClassifier

x=df[['XX_Women_Who_Are_Aware_Of_Danger_Signs_Of_Ari_Pneumonia_Total','YY_Crude_Death_Rate_Cdr_Total_Person','MM_Pregnancy_To_Women_Aged_15_49_Years_Resulting_In_Abortion_Total']]
y=df['AA_Population_Total']

lin_reg = LinearRegression()
print(lin_reg.fit(x, y))
print("Linear Regression accuracy: ", (lin_reg.score(x,y))*100)

Rndm_modl = RandomForestClassifier()
Rndm_modl.fit(x,y)
print("Random Forest accuracy: ", (Rndm_modl.score(x,y))*100)

decs_modl = DecisionTreeClassifier()
decs_modl.fit(x,y)
print("Decision Tree Classifier accuracy: ", (decs_modl.score(x,y))*100)

svm_modl = SVC()
svm_modl.fit(x,y)
print("SVM accuracy: ", (svm_modl.score(x,y))*100)
```
